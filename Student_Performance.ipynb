```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
df = pd.read_csv("student_data.csv")
df.head()
```

```python
#Inspect data types and missing values
df.info()
```

```python
#Statistical summary
df.describe()
```

```python
#Count null values in each column
df.isnull().sum()
```

```python
#Count duplicate rows
print(f"Duplicate rows found: {df.duplicated().sum()}")
```

This confirmed the dataset is complete with 0 null entries and 0 duplicates. No deletion or imputation is required.

```python
#Check columns
df.columns
```

```python
#Make all column names lower case
df.columns = df.columns.str.lower()
```

```python
#Check for possible typos
print(df['school'].unique())
print(df['fjob'].unique())
print(df['mjob'].unique())
```

```python
#Percentage breakdown of which school students attend
(df['school'].value_counts(normalize=True).round(4) * 100).astype(str)+"%"
```

```python
#Flag outliers
numerical_col = df.select_dtypes(include='number').columns.tolist()
outliers = {}
for col in numerical_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)

    outliers[col] = df.loc[(df[col] < lower_bound)|(df[col] > upper_bound), col]
    print(f"\nOutliers in {col}:\n", outliers[col])
```

```python
#Visualize high variance outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['absences'], color='skyblue')
plt.title("'Absences' Outliers")
plt.show()
```

```python
#Distribution of G3 (Final Grade)
plt.figure(figsize=(6, 4))
sns.histplot(df['g3'], kde=True, color='skyblue')
plt.title("Final Grade (G3) Distribution")
plt.ylabel("# of students")
plt.show()
```

```python
#Study time vs G3
plt.figure(figsize=(6, 4))
sns.regplot(x=df['studytime'], y=df['g3'], color='skyblue')
plt.title("Study Time vs Final Grade")
plt.show()
```

There is a slight positive correlation between study time and final grade but it is not as strong as expected.

```python
#Absences vs G3
plt.figure(figsize=(6, 4))
sns.regplot(x=df['absences'], y=df['g3'], color='skyblue')
plt.title("Absences vs Final Grade")
plt.show()
```

Higher absences unexpectedly led to sightly higher grades, which contradicts the common sense. This is likely distorted by outliers.

```python
#Try removing extreme outliers
df_filtered = df[df['absences'] <= 40]

plt.figure(figsize=(6, 4))
sns.regplot(x=df_filtered['absences'], y=df_filtered['g3'], color='skyblue')
plt.title("Absences vs Final Grade")
plt.show()
```

```python
#Verify with correlation
df[['absences', 'g3']].corr()
```

Although the regression line shows a slight positive slope, the effect size is extremely small. This shows that absences have minimal practical impact on final grades in this dataset (verified with correlation).

```python
#Failures vs G3
plt.figure(figsize=(6, 4))
sns.regplot(x=df['failures'], y=df['g3'], color='skyblue')
plt.title("Past Failures vs Final Grade")
plt.show()
```

As the number of past failures increases, the student's final grade tends to significantly decrease.

```python
#G1, G2, G3 correlation
df[['g1','g2','g3']].corr()
```

This shows a very strong positive correlation between all the grades, specifically G2 being the strongest predictor for G3.

```python
#New column called 'improvement'
df['improvement'] = (df['g3'] - df['g1'])
df['improvement']
```

```python
df['performance'] = "passed"

df.loc[df['g3'] < 10, 'performance'] = "no_change"
df.loc[(df['g3'] < 10) & (df['improvement'] <= 0), 'performance'] = "declined"
df.loc[(df['g3'] < 10) & (df['improvement'] >= 2), 'performance'] = "improved"

sns.countplot(x=df['performance'], color='skyblue')
plt.title("Breakdown of Students' Performance")
plt.xlabel("category")
plt.ylabel("# of students")
plt.show()
```

**Final Key Insights**
1. G2 is the strongest predictor of G3.
2. Study time has a mild positive impact.
3. Absences show very weak correlation after removing outliers.
4. Past failures have significant negative correlation on G3.
5. Parental education may have moderate influence.
